```python
conf_df = pd.read_csv('cases - confirmed - jhu.csv')
deaths_df = pd.read_csv('cases - deaths - jhu.csv')
recv_df = pd.read_csv('cases - recovered - jhu.csv')
```

```python
conf_df = pd.read_csv('time_series_2019-ncov-Confirmed.csv')
deaths_df = pd.read_csv('time_series_2019-ncov-Deaths.csv')
recv_df = pd.read_csv('time_series_2019-ncov-Recovered.csv')
```

```python
dates = ['1/21/20 22:00',
       '1/22/20 12:00', '1/23/20 12:00', '1/24/20 0:00', '1/24/20 12:00',
       '1/25/20 0:00', '1/25/20 12:00', '1/25/20 22:00', '1/26/20 11:00',
       '1/26/20 23:00', '1/27/20 9:00', '1/27/20 19:00', '1/27/20 20:30',
       '1/28/20 13:00', '1/28/20 18:00', '1/28/20 23:00', '1/29/20 13:30',
       '1/29/20 14:30', '1/29/20 21:00', '1/30/20 11:00', '1/31/20 14:00',
       '2/1/20 10:00', '2/2/20 21:00', '2/3/20 21:00', '2/4/20 9:40',
       '2/4/20 22:00', '2/5/20 9:00', '2/5/20 23:00', '2/6/20 9:00',
       '2/6/20 14:20', '2/7/20 20:13', '2/7/20 22:50', '2/8/20 10:24',
       '2/8/20 23:04', '2/9/20 10:30', '2/9/20 23:20', '2/10/20 10:30',
       '2/10/20 19:30', '2/11/20 10:50', '2/11/20 20:44', '2/12/20 10:20',
       '2/12/20 22:00']
```

```python
conf_df_long = conf_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Confirmed')

deaths_df_long = deaths_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Deaths')

recv_df_long = recv_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Recovered')

full_table = pd.concat([conf_df_long, deaths_df_long['Deaths'], recv_df_long['Recovered']], axis=1, sort=False)
```

```python
full_table.shape
```

```python
pd.to_datetime?
```

```python
full_table['Date'] = pd.to_datetime(full_table['Date'])
```

```python
# filling missing values with 0 in columns ('Confirmed', 'Deaths', 'Recovered')
full_table[['Confirmed', 'Deaths', 'Recovered']] = full_table[['Confirmed', 'Deaths', 'Recovered']].fillna(0)
```

```python
# selecting only the last entry for each day 
full_table = full_table[full_table['Date'].eq(full_table.groupby([full_table['Date'].dt.year,
                                                     full_table['Date'].dt.month, 
                                                     full_table['Date'].dt.day])['Date'].transform('max'))]
```

```python
# changing datetime to just date
full_table['Date'] = pd.to_datetime(full_table['Date']).dt.date
```

```python
# adding two more columns
full_table['Deaths to Confirmed'] = round(full_table['Deaths']/full_table['Confirmed'], 3)
full_table['Recovered to Confirmed'] = round(full_table['Deaths']/full_table['Confirmed'], 3)

full_table = full_table[full_table['Province/State']!='Diamond Princess cruise ship']

full_table.head()
```
